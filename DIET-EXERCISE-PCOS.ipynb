Does PCOS affect mental health?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot for Mental Health vs PCOS
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Mental_Health', hue='PCOS', palette='Set2')
plt.title('Effect of PCOS on Mental Health', fontsize=16)
plt.xlabel('Mental Health Issues', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='PCOS', labels=['No', 'Yes'], fontsize=12)
plt.show()

# Cross-tabulation for numerical insights
mental_health_vs_pcos = pd.crosstab(df['Mental_Health'], df['PCOS'])
mental_health_vs_pcos_percentage = mental_health_vs_pcos.div(mental_health_vs_pcos.sum(axis=1), axis=0) * 100

mental_health_vs_pcos, mental_health_vs_pcos_percentage
```

Do PCOS only impact non-married females?


```python
# Bar plot: Marital Status vs. PCOS Status
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Marital_Status_binary', hue='PCOS_binary', palette="coolwarm")
plt.title("PCOS Status by Marital Status ( 0.0 = Unmarried, 1.0 = Married)", fontsize=14)
plt.xlabel("Marital Status",fontsize=12)
plt.ylabel("Female count", fontsize=12)
plt.legend(title="PCOS Status", labels=["No", "Yes"])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group data by Vegetarian status and PCOS status and count the occurrences
vegetarian_pcos_count = (
    df.groupby(['Vegetarian', 'PCOS_binary'])
    .size()
    .reset_index(name='Count')
)

# Bar plot: Vegetarian status vs. PCOS counts
plt.figure(figsize=(10, 6))
sns.barplot(
    data=vegetarian_pcos_count, 
    x='Vegetarian', 
    y='Count', 
    hue='PCOS_binary', 
    palette="muted"
)
plt.title("Vegetarians with and without PCOS", fontsize=14)
plt.xlabel("Vegetarian Status (No/Yes)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="PCOS Status (0 = No, 1 = Yes)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group and count occurrences of Stress Level by Exercise Frequency and Benefit
exercise_stress = (
    df.groupby(['Exercise_Frequency', 'Exercise_Benefit', 'Stress_Level'])
    .size()
    .reset_index(name='Count')
)

# Pivot for heatmap structure
exercise_stress_pivot = exercise_stress.pivot_table(
    index='Exercise_Frequency',
    columns='Exercise_Benefit',
    values='Count',
    aggfunc='sum',
    fill_value=0
)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    exercise_stress_pivot,
    annot=True,
    cmap="YlGnBu",
    fmt='d',
    cbar_kws={'label': 'Count'}
)
plt.title("Exercise Frequency, Benefit, and Stress Level", fontsize=16)
plt.xlabel("Exercise Benefit", fontsize=12)
plt.ylabel("Exercise Frequency", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Group data by Sleep Hours and calculate the average Stress Level
sleep_stress = (
    df.groupby('Sleep_Hours')['Stress_Level']
    .value_counts(normalize=True)  # Alternatively, use `.mean()` for average
    .reset_index(name='Proportion')
)

# Line plot of Sleep Hours vs. Stress Level
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=sleep_stress,
    x='Sleep_Hours',
    y='Proportion',
    hue='Stress_Level',
    marker='o',
    palette="viridis"
)
plt.title("Stress Level Across Different Sleep Hours", fontsize=16)
plt.xlabel("Sleep Hours", fontsize=12)
plt.ylabel("Proportion", fontsize=12)
plt.legend(title="Stress Level")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
